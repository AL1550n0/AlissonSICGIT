# Actividad: Ordenamiento de Burbuja


## Instrucciones

1. Explique brevemente y en forma clara cómo funciona el algoritmo de ordenamiento de burbuja.

2. Desarrollo un función donde se use el ordenamiento de burbuja aplicada a una lista de valores enteros.



```python

#Funcionamiento del algoritmo de ordenamiento de burbuja
#Compara dos elementos en una lista si el elemento de la izquiera es mayor lo intercambia con la derecha. Se repite varias veces hasta
# que el elemento de mayor tamaño pase al final de la lista.

def ordenar_estudiantes_por_edad(estudiantes):
    n = len(estudiantes)
    # Recorrer todos los elementos de la lista
    for i in range(n - 1):
        # Últimos i elementos ya están ordenados
        for j in range(n - i - 1):
            if estudiantes[j][1] > estudiantes[j + 1][1]:
                # Intercambiar si la edad actual es mayor que la siguiente
                estudiantes[j], estudiantes[j + 1] = estudiantes[j + 1], estudiantes[j]
    return estudiantes

# Ejemplo de uso: lista de tuplas (nombre, edad)
estudiantes = [("Alexis", 19), ("Carlos", 21), ("Alisson", 18), ("Alejandra", 22), ("Josue", 20), ("Israel", 17), ("Enrique", 23)]
estudiantes_ordenados = ordenar_estudiantes_por_edad(estudiantes)
print("Nombre de los estudiantes ordenados por edad:", estudiantes_ordenados)
```

# Actividad: Función Swap.

1. Escriba una función que permita el intercambion de elementos en una lista.
2. Los argumentos de esta función serán, la lista y los indices de los elementos que se desean intercambiar.
3. En la siguiente celda desarrollo el código requerido, recuerde agregar comentarias breves que expliquen su funcionamiento.


```python
# Espacio para desarrollar la actividad, si tienes alguna duda o comentario adicional
# lo puedes dejar planteado acá.

def intercambiar(lista, i, j):
    """
    Intercambia dos elementos en una lista dados sus índices.

    Parámetros:
    lista (list): Lista de elementos.
    i (int): Índice del primer elemento a intercambiar.
    j (int): Índice del segundo elemento a intercambiar.

    Retorna:
    list: La lista con los elementos intercambiados.
    """
    # Verificar que los índices estén dentro del rango de la lista
    if 0 <= i < len(lista) and 0 <= j < len(lista):
        # Intercambiar los elementos utilizando asignación múltiple
        lista[i], lista[j] = lista[j], lista[i]
    else:
        raise IndexError("Uno o ambos índices están fuera del rango de la lista.")
    return lista

# Ejemplo 
personas = [
    {"nombre": "Luis", "edad": 25},
    {"nombre": "Mateo", "edad": 30},
    {"nombre": "Susana", "edad": 22},
    {"nombre": "Alejandro", "edad": 28}
]

print("Lista sin intercambiar datos:")
print(personas)

# Intercambiar los elementos en los índices 2 y 3
nueva_lista = intercambiar(personas, 2, 3)

print("\nLista con los datos intercambiados:")
print(nueva_lista)
```

```python
# Celda para desarrollar 


def quick_sort(list):
    """
    Implementación del algoritmo Quick Sort. 
    Este es un algoritmo que utiliza el divide y vencerás. Escoge un elemento como pivote
    la lista se divide en sublistas y se vuelve recursivo.

    Parámetros:
    lista (list): La lista a ordenar.

    Retorna:
    list: Lista ordenada.
    """
    # Caso base: listas de longitud 0 o 1 ya están ordenadas
    if len(list) <= 1:
        return list
    
    # Elegir el pivote (en este caso, el último elemento)
    pivote = list[-1]
    
    # Sublistas para menores y mayores al pivote
    menores = [x for x in list[:-1] if x <= pivote]
    mayores = [x for x in list[:-1] if x > pivote]
    
    # Llamada recursiva y concatenación
    return quick_sort(menores) + [pivote] + quick_sort(mayores)

# Ejemplo de uso
list = [10, 12, 16, 17, 11, 15, 13, 14]
print("Lista original:", list)
print("Lista ordenada:", quick_sort(list))
```

# Actividad: Quick Sort usando un valor de pivote aleatorio.

1. Explique brevemente el funcionamiento del algoritmo Quick Sort en la modaladidad de usar un valor de pivote aleatorio.
2. Comente las ventajas de esta variación.
3. Desarrollo el código necesarion para este algoritmo.
4. Agregue comentarios sobre su funcionamiento.



```python
# Espacio para desarrollar el código siguiendo las instrucciones.


from random import randint

def partition2(S, low, high):
    """
    Selecciona un pivote aleatorio y reorganiza
    los elementos para que los menores o iguales queden a la izquierda,y los mayores a la derecha.
    Esta función evita la probabilidad de que el algoritmo caiga en su peor caso.

    Ventajas:
    Una mejora en las distribuciones de las particiones
    Tiene una compplejidad promedio de O(nlogn)

    Parámetros:
    S (list): Lista a ordenar.
    low (int): Índice inferior.
    high (int): Índice superior.

    Return:
    int: Índice final del pivote.
    """
    # Seleccionar un pivote aleatorio
    rand = randint(low, high)
    S[low], S[rand] = S[rand], S[low]  # Intercambiar pivote aleatorio con el primer elemento
    pivot, left, right = S[low], low, high
    
    print(S, left, right, "pivot =", pivot)  # Para visualizar las particiones
    
    while left < right:
        # Avanzar left hasta encontrar un valor mayor al pivote
        while left < high and S[left] <= pivot:
            left += 1
        # Retroceder right hasta encontrar un valor menor al pivote
        while right > low and S[right] > pivot:
            right -= 1
        # Si left aún es menor que right, intercambiar valores
        if left < right:
            S[left], S[right] = S[right], S[left]
    
    # Colocar el pivote en su posición correcta
    S[low], S[right] = S[right], S[low]
    
    return right

def quicksort2(S, low, high):
    """
    Implementación del algoritmo Quick Sort utilizando partition2.

    Parámetros:
    S (list): Lista a ordenar.
    low (int): Índice inferior.
    high (int): Índice superior.
    """
    if low < high:
        # Obtener índice del pivote tras partición
        pivot_index = partition2(S, low, high)
        # Ordenar sublistas izquierda y derecha
        quicksort2(S, low, pivot_index - 1)
        quicksort2(S, pivot_index + 1, high)

# Lista de ejemplo
S = [15, 10, 12, 20, 25, 13, 22]
quicksort2(S, 0, len(S) - 1)
print("Lista ordenada:", S)
```
