## Q 27-01

¿Cuántas comparaciones se ejecutaron en el siguiente proceso de ordenamiento de burbuja?
```
def bubbleSort(S):
    n = len(S)
    for i in range(n):
        print(S)
        for j in range(n-1):
            if S[j] > S[j+1]:
               S[j] , S[j+1] = S[j+1] , S[j]

S = [50, 30, 40, 10, 20]
bubbleSort(S)
print(S)
```
#### Output
```
[50, 30, 40, 10, 20]
[30, 40, 10, 20, 50]
[30, 10, 20, 40, 50]
[10, 20, 30, 40, 50]
[10, 20, 30, 40, 50]
[10, 20, 30, 40, 50]
```

En la primer pasada se hicieron (n-1) comparaciones, o sea 4.
En la segunga, el 50 es el mayor y ya está en su posición, entonces se realizaron 3 comparaciones. 
En la tercer pasada se realizaron 2 comparaciones. 
En la cuarta pasada se realiza 1 comparación,
En total son 10 comparaciones.

## Q 27-02

```
def insertionSort2(S):
    n = len(S)
    for i in range(1,n):
        print(S)
        x = S[i]
        j = i - 1
        while j >= 0 and S[j] > x:
            S[j+1] = S[j]
            j -= 1
        S[j + 1] = x

S = [50, 30, 40, 10, 20]
insertionSort2(S)
print(S)
```
#### Output
```
[50, 30, 40, 10, 20]
[30, 40, 10, 20, 50]
[30, 10, 20, 40, 50]
[10, 20, 30, 40, 50]
[10, 20, 30, 40, 50]
[10, 20, 30, 40, 50]
```

````python
def insertionSort2(S):
    n = len(S)
    comparisons = 0
    for i in range(1, n):
        print(S)
        x = S[i]
        j = i - 1
        while j >= 0:
            comparisons += 1  # Contamos la comparación, verdadera o falsa
            if S[j] > x:
                S[j+1] = S[j]
                j -= 1
            else:
                break
        S[j + 1] = x
    print(f"Total comparisons: {comparisons}")
    return S

S = [50, 30, 40, 10, 20]
insertionSort2(S)
print(S)
````

Se realizan un total de 10 comparaciones si se cuentan las comparaciones en las que el número de la derecha es menor que el de la izquierda, sino serían 8. 
En la primer iteración solo hay un cambio.
En la segunda iteración se comparan 2 pero solo hay 1 cambio. Son 2 cambios pero 3 comparaciones.
En la tercer iteración se 3 y se realizan 3 cambios, en total Son ya 6 comparaciones. Pero 5 acumuladas.
En la cuarta iteración se comparan 4 pero solo se realizan 3 cambios. Van 10 comparaciones pero solo 8 acumuladas.
